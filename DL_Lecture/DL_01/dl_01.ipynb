```python
import numpy as np

def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i] - mx) * (y[i] - my)
  return d

def predict(x):
  return fake_a * x + fake_b

def mse(y, y_pred):
  return (1/n) * sum((y - y_pred)**2)

fake_a = 3
fake_b = 76

x = np.array([2, 4, 6, 8])
y = np.array([81,93,91,97])

n = len(x)

mx = np.mean(x)
my = np.mean(y)

print("mean of X", mx)
print("mean of Y", my)
```

```python
divisor = sum([(i - mx)**2 for i in x])

dividend = top(x, mx, y, my)

print("분모: ", divisor)
print("분자: ", dividend)
```

```python
a = dividend/ divisor

b = my - (mx*a)


print("기울기 a = ", a)
print("y 절편 b = ", b)
```

```python
predict_result = []

for i in range(n):
  predict_result.append(predict(x[i]))
  print("공부시간 = %.f, 실제점수 = %.f, 예측점수 = %.f" % (x[i], y[i], predict(x[i])))
```

```python
print("평균 제곱 오차: " + str(mse(y, predict_result)))
```
