```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/jbrownlee/Datasets/master/ionosphere.csv", header=None)

df[34] = df[34].map({'g': 0, 'b': 1})

X = df.iloc[:, 0:34]
y = df.iloc[:,34]
```

```python
model = Sequential()
model.add(Dense(60, input_dim=34, activation='relu', name='Dense_1'))
model.add(Dense(30, activation='relu', name='Dense_2'))
model.add(Dense(1, activation='sigmoid', name='Dense_3'))
model.summary()

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(X, y, epochs=350, batch_size=35)
```
